```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
file_one = "Resources/2022-City-wide-Emissions.csv"
```

```python
# Read our Data file with the pandas library
city_emissions_df = pd.read_csv(file_one, encoding="ISO-8859-1")
```

```python
# Show just the header
city_emissions_df.head()
```

```python
city_emissions_df.columns
```

```python
cleaned_emission_df = city_emissions_df.dropna(how='any')
cleaned_emission_df.head()
```

```python
renamed_emission_df = cleaned_emission_df.rename(columns={'Gases Included': 'Gases', 'Emissions Column Number': 'Emissions_Column_Number',
                                                          'Emissions Column Name': 'Emissions_Column_Name', 'Emissions Row Number': 'Emissions_Row_Number',
                                                          'Emissions Row Name': 'Emissions_Row_Name', 'Emissions Response Answer': 'Emissions_Response_Answer',
                                                          'Emissions Description':'Emissions_Description', 'Emissions Data Reported': 'Emissions_Data_Reported',
                                                         'Inventory Year': 'Inventory_Year', 'Population in inventory year':'Population',
                                                         'Last update':'Last_update', 'Organization Name': 'Organization_Name'})
US_cities_emissions = renamed_emission_df.groupby(['Country', 'City', 'City_Location', 'Inventory_Year']).max('Emissions_Response_Answer').reset_index()

US_cities_emissions['City_Location'] = US_cities_emissions['City_Location'].str.replace("POINT ", "")
US_cities_emissions.head(100)
```

```python
US_cities_emissions[['latitude', 'longitude']] = US_cities_emissions['City_Location'].str.split(' ', 1, expand=True)
US_cities_emissions['latitude'] = US_cities_emissions['latitude'].str.replace('(', '').astype(str).astype(float)
US_cities_emissions['longitude'] = US_cities_emissions['longitude'].str.replace(')', '').astype(str).astype(float)
final_US_cities_emissions =US_cities_emissions[['Country','City','latitude','longitude','Inventory_Year',
                    'Emissions_Column_Number', 'Emissions_Row_Number', 
                    'Emissions_Response_Answer', 'Emissions_Data_Reported','Population']]
final_US_cities_emissions.head()
```

```python
final_US_cities_emissions.dtypes
```

```python
# Export reduced_emission_df as CSV files.
final_US_cities_emissions.to_csv("Output/us_city_emissions_df.csv", index=False)
```
